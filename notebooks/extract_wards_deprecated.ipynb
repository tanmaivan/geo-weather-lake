## Extract Wards for Quảng Ngãi (Deprecated)

## This process is no longer part of the active pipeline and is retained only for traceability.

Purpose: Filter data from the full Vietnam dataset (VietnamWardBoundary2025.csv) to get the list of wards/communes in Quảng Ngãi province:

- Read the original CSV file.
- Provide basic statistics and unique values for columns.
- Filter records where tentinh equals “tỉnh Quảng Ngãi”.
- Extract the tenhc column (ward/commune names) and save to quang_ngai_wards.csv.

### 1. Setup


```python
import pandas as pd 
from pathlib import Path
```

### 2. Load data


```python
project_root = Path.cwd()
input_csv_path = project_root / Path("..") / "resources" / "VietnamWardBoundary2025.csv"
output_csv_path = project_root / Path("..") / "resources" / "quang_ngai_wards.csv"

print(f"Input file path: {input_csv_path.resolve()}")
print(f"Output file path: {output_csv_path.resolve()}")
```

```python
df = pd.read_csv(input_csv_path)
df.head()
```

```python
df.nunique()
```

```python
df.info()
```

### 3. Filter province and extract wards


```python
target_province = 'tỉnh Quảng Ngãi'
df_qn = df[df['tentinh'] == target_province].copy()
len(df_qn)
```

```python
df_wards = df_qn[['tenhc']]
df_wards
```

### 4. Save output


```python
df_wards.to_csv(output_csv_path, index=False)
```
